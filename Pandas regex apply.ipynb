```python
import pandas as pd
```

```python
nmah_df = pd.read_csv('nmah_df.tsv', sep='\t')
nmah_df.head()
```

```python
def split_record_id(row):
    record_id = row['record_ID']
    record_split = record_id.split('_')
    just_the_id = record_split[1]
    row['just_the_id'] = just_the_id
    return row
```

```python
new_df = nmah_df.apply(split_record_id, axis=1)
new_df.head()
```

```python
import re
```

```python
def extract_years(row):
    four_number_pattern = r'^(\d{4})$'
    date_check = re.match(four_number_pattern, row['Date'])
    if date_check:
        row['start_year'] = date_check.group(1)
        row['end_year'] = date_check.group(1)
    date_range_pattern = r'^(\d{4})\s*-\s*(\d{4})$'
    date_check = re.match(date_range_pattern, row['Date'])
    if date_check:
        row['start_year'] = date_check.group(1)
        row['end_year'] = date_check.group(2)    
    return row
```

```python
date_df = nmah_df.fillna('').apply(extract_years, axis = 1)
date_df.head()
```
