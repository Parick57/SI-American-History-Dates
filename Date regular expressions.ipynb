```python
import pandas as pd
```

```python
nmah_df = pd.read_csv('nmah_df.tsv', sep='\t')
nmah_df.head()
```

```python
pattern = r'^\d'
nmah_df['Date'].str.match(pattern).fillna(False)
```

```python
nmah_df[nmah_df['Date'].str.match(pattern).fillna(False)]
```

```python
pattern = r'^\d{4}$'
nmah_df[nmah_df['Date'].str.match(pattern).fillna(False)]
```

```python
nmah_df[pd.isnull(nmah_df['Date'])]
```

```python
four_number_pattern = r'^\d{4}$'
four_number = nmah_df['Date'].str.match(four_number_pattern).fillna(False)
null_filter = pd.isnull(nmah_df['Date'])
nmah_df[~(four_number | null_filter)]
```

```python
date_range_pattern = r'^\d{4}\s*-\s*\d{4}$'
nmah_df[nmah_df['Date'].str.match(date_range_pattern).fillna(False)]
```

```python
date_range_filter = nmah_df['Date'].str.match(date_range_pattern).fillna(False)
nmah_df[~(four_number | null_filter | date_range_filter)]
```
