```python
import pandas as pd
```

```python
nmah_df = pd.read_csv('nmah_df.tsv', sep='\t')
nmah_df.head()
```

```python
pattern = r'^\d'
nmah_df['Date'].str.match(pattern).fillna(False)
```

```python
nmah_df[nmah_df['Date'].str.match(pattern).fillna(False)]
```

```python
pattern = r'^\d{4}$'
nmah_df[nmah_df['Date'].str.match(pattern).fillna(False)]
```

```python
nmah_df[pd.isnull(nmah_df['Date'])]
```

```python
four_number_pattern = r'^\d{4}$'
four_number = nmah_df['Date'].str.match(four_number_pattern).fillna(False)
null_filter = pd.isnull(nmah_df['Date'])
nmah_df[~(four_number | null_filter)]
```

```python
date_range_pattern = r'^\d{4}\s*-\s*\d{4}$'
nmah_df[nmah_df['Date'].str.match(date_range_pattern).fillna(False)]
```

```python
date_range_filter = nmah_df['Date'].str.match(date_range_pattern).fillna(False)
nmah_df[~(four_number | null_filter | date_range_filter)]
```

```python
after_pattern = r'^[Aa]fter\s\d{4}$'
nmah_df[nmah_df['Date'].str.match(after_pattern).fillna(False)]
```

```python
before_pattern = r'^[Bb]efore\s\d{4}$'
nmah_df[nmah_df['Date'].str.match(before_pattern).fillna(False)]
```

```python
circa_pattern = r'^c|ca\s\d{4}$'
nmah_df[nmah_df['Date'].str.match(circa_pattern).fillna(False)]
```

```python
date_range_filter = nmah_df['Date'].str.match(date_range_pattern).fillna(False)
after_filter = nmah_df['Date'].str.match(after_pattern).fillna(False)
before_filter = nmah_df['Date'].str.match(before_pattern).fillna(False)
circa_filter = nmah_df['Date'].str.match(circa_pattern).fillna(False)
nmah_df[~(four_number | null_filter | date_range_filter | after_filter | before_filter | circa_filter)]
```

```python
s_pattern = r'^\d{4}[s]s*-\s*\d{4}[s]$'
nmah_df[nmah_df['Date'].str.match(s_pattern).fillna(False)]
```

```python
sapos_pattern = r'^\d{4}[s]s*-\s*\d{4}[s]$'
nmah_df[nmah_df['Date'].str.match(sapos_pattern).fillna(False)]
```

```python
date_range_filter = nmah_df['Date'].str.match(date_range_pattern).fillna(False)
after_filter = nmah_df['Date'].str.match(after_pattern).fillna(False)
before_filter = nmah_df['Date'].str.match(before_pattern).fillna(False)
circa_filter = nmah_df['Date'].str.match(circa_pattern).fillna(False)
s_filter = nmah_df['Date'].str.match(s_pattern).fillna(False)
nmah_df[~(four_number | null_filter | date_range_filter | after_filter | before_filter | circa_filter  | s_filter)]
```
