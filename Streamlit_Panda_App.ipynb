```python
nmah_df = pd.read_csv('nmah_df.tsv', sep='\t')
nmah_df.head()
```

```python
def split_date(row):
    date = str(row['Date'])
    record_split = date.split('-')
    if len(record_split) > 1:
        what_is_the_date = record_split[1]
        row['what_is_the_date'] = what_is_the_date
    return row
```

```python
new_df = nmah_df.apply(split_date, axis=1)
new_df.head()
```

```python
def extract_years(row):
    four_number_pattern = r'^(\d{4})$'## Four number date Ex.2000
    date_check = re.match(four_number_pattern, row['Date'])
    if date_check:
        row['start_year'] = date_check.group(1)
        row['end_year'] = date_check.group(1)
    date_range_pattern = r'^(\d{4})\s*-\s*(\d{4})$' ## Date - Date Ex. 2000 - 2001
    date_check = re.match(date_range_pattern, row['Date'])
    if date_check:
        row['start_year'] = date_check.group(1)
        row['end_year'] = date_check.group(2)    
    return row
```

```python
date_df = nmah_df.fillna('').apply(extract_years, axis = 1)
date_df.head()
```
